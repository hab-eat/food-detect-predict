```python
from ultralytics import YOLO  # type: ignore
model = YOLO("runs/classify/train6/weights/best.pt")
results = model(
    'https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcTwyHp2p-x5_o7RUU-cxlg_8JD5iOBvIdXy7w&s')
for result in results:
    probs = result.probs
    class_index = probs.top1
    class_name = result.names[class_index]
    score = float(probs.top1conf)

    print(probs)
    print(class_index)
    print(class_name)
    print(score)
```

```python
metrics = model.val()
metrics.top1
metrics.top5
print(metrics)
```
